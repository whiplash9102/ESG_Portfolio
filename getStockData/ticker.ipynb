```python
import refinitiv.data as rd
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import yfinance as yf
```

```python
rd.open_session()
```

```python
tickers = [
    "ASML.AS",
    "MC.PA",
    "SAN.PA",
    "AIR.PA",
    "SU.PA",
    "OR.PA",
    "BNP.PA",
    "RMS.PA",
    "DG.PA",
    "ADS.DE"
]
```

```python
data = []
for t in tickers: 
    info = yf.Ticker(t).info
    data.append({
        "Ticker": t,
        "Name": info.get("longName"),
        "Sector": info.get("sector"),
        "Industry": info.get("industry")
    })

df_meta = pd.DataFrame(data)
print(df_meta)
df_meta.to_excel("company_info.xlsx", index=False)
```

```python
end = datetime.today()
start = end - timedelta(days=365)
```

### Get historical price data for the selected stocks

```python
prices = yf.download(
    tickers,
    start="2022-01-01",
    end="2026-01-01"
)["Close"]
```
